```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    
    def reverseList(self, head):
        """
        Reverses a singly linked list
        """
        temp = head
        prev = None

        if head is None or head.next is None:
            return head

        while temp is not None:
            next = temp.next
            temp.next = prev
            prev = temp  # This was the typo (corrected from `perv` to `prev`)
            temp = next

        return prev  # Return the new head of the reversed list

    def findKth(self, temp, k):
        """
        Finds the k-th node from the current node (1-based index)
        """
        k -= 1  # Adjust to 0-based indexing
        while temp is not None and k > 0:
            k -= 1
            temp = temp.next
        return temp  # Returns the k-th node or None if the list is shorter than k nodes

    def reverseKGroup(self, head, k):
        """
        Reverses nodes in k groups. If the number of nodes is not a multiple of k, the final group remains as is.
        """
        temp = head
        prev = None

        # Handle edge case if the list is empty
        if head is None or k == 1:
            return head

        while temp is not None:
            # Find the k-th node from the current position
            kth = self.findKth(temp, k)
            if kth is None:  # If less than k nodes remain, break
                if prev:
                    prev.next = temp
                break

            # Save the next group to process
            next = kth.next
            kth.next = None  # Temporarily break the list for reversal

            # Reverse the current group
            reversedHead = self.reverseList(temp)

            # Connect the previous part to the newly reversed list
            if prev is None:
                head = reversedHead  # If this is the first group, update head
            else:
                prev.next = reversedHead

            # The new `prev` is the original `temp` (start of the group before reversal)
            prev = temp

            # Move to the next group
            temp = next

        return head
```
