```python
from collections import deque

class Solution:
    def shortestPath(self, edges, n, m, src):
        # Step 1: Create adjacency list
        adj = [[] for _ in range(n)]
        
        for i in range(m):
            u = edges[i][0]
            v = edges[i][1]
            adj[u].append(v)
            adj[v].append(u)  # Assuming the graph is undirected, add both directions
        
        # Step 2: Initialize visited and distance arrays
        dist = [-1] * n  # Use -1 to indicate that the node is not reachable
        dist[src] = 0  # Distance to the source node is 0
        
        # Step 3: BFS to find shortest paths
        q = deque([src])  # Queue for BFS
        
        while q:
            node = q.popleft()
            
            # Visit all neighbors
            for neighbor in adj[node]:
                if dist[neighbor] == -1:  # Not visited
                    dist[neighbor] = dist[node] + 1  # Update the shortest distance
                    q.append(neighbor)
        
        return dist
```

```python
from typing import List

class Solution:

    def shortestPath(self, V: int, E: int, edges: List[List[int]]) -> List[int]:
        
        # Initialize adjacency list
        adj = [[] for _ in range(V)]
        st = []
        vis = [0] * V
                         
        # Build adjacency list from the edges list
        for i in range(E):
            u = edges[i][0]
            v = edges[i][1]
            d = edges[i][2]
            adj[u].append([v, d])
        
        # Helper function to perform DFS for topological sort
        def dfs(node):
            vis[node] = 1
            for i in adj[node]:
                n = i[0]
                if vis[n] == 0:
                    dfs(n)
            st.append(node)
        
        # Perform DFS for all nodes to get topological order
        for i in range(V):
            if vis[i] == 0:
                dfs(i)
        
        # Initialize distance array with infinity
        dist = [float('inf')] * V
        dist[0] = 0  # Assuming 0 as the source
        
        # Process nodes in topological order
        while st:
            node = st.pop()
            
            # Only proceed if the current node's distance is not infinity
            if dist[node] != float('inf'):
                for i in adj[node]:
                    n = i[0]
                    w = i[1]
                    
                    # Update distance if a shorter path is found
                    if dist[node] + w < dist[n]:
                        dist[n] = dist[node] + w
        for i in range(V):
            if dist[i] == float('inf'):
                dist[i] = -1

        return dist
```
