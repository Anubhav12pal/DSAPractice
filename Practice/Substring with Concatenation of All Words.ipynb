```python
class Solution(object):
    def condi(self, l, s, words, word_len, required_len):
        # Build a frequency map for the words
        word_count = {}
        for word in words:
            word_count[word] = 1 + word_count.get(word, 0)

        # Validate the substring
        for i in range(l, l + required_len, word_len):
            word = s[i:i + word_len]
            if word in word_count:
                word_count[word] -= 1
                if word_count[word] == 0:
                    del word_count[word]
            else:
                return False

        # If all words matched, the frequency map should be empty
        return not word_count

    def findSubstring(self, s, words):
        """
        :type s: str
        :type words: List[str]
        :rtype: List[int]
        """
        if not s or not words:
            return []

        word_len = len(words[0])  # Length of each word
        num_words = len(words)   # Number of words
        required_len = word_len * num_words  # Total length of concatenated substring
        n = len(s)
        result = []

        for i in range(n - required_len + 1):
            if self.condi(i, s, words, word_len, required_len):
                result.append(i)

        return result
```

```python
class Solution:
    def findSubstring(self, s, words):
        if not s or not words:
            return []
        
        word_len = len(words[0])
        word_count = len(words)
        substring_len = word_len * word_count
        n = len(s)
        
        # Frequency map of words
        word_map = {}
        for word in words:
            word_map[word] = 1 + word_map.get(word, 0)
        
        result = []

        # Iterate for each starting index
        for i in range(word_len):
            left = i
            right = i
            curr_map = {}
            count = 0  # Count of matched words
            
            # Sliding window
            while right + word_len <= n:
                word = s[right:right + word_len]
                right += word_len
                
                # Check if the word is valid
                if word in word_map:
                    curr_map[word] = 1 + curr_map.get(word, 0)
                    count += 1
                    
                    # Remove extra words
                    while curr_map[word] > word_map[word]:
                        left_word = s[left:left + word_len]
                        curr_map[left_word] -= 1
                        left += word_len
                        count -= 1
                    
                    # Valid substring
                    if count == word_count:
                        result.append(left)
                else:
                    # Reset window if word is invalid
                    curr_map.clear()
                    count = 0
                    left = right

        return result
```
